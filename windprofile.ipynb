```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

df = pd.read_csv('data2-2.txt', sep=r'\s+', skiprows=5)
df = df.apply(lambda x: x.str.replace(',','.'))
df = df.astype(float)
```

```python
df.dtypes
```

```python
# add timeframe at 20Hz
ts = np.arange(0,len(df),1/20)
ts = pd.DataFrame(ts)
df['f(Hz)'] = ts
df = df.set_index('f(Hz)')
```

```python
vel_vec = ['u','v','w']
```

```python
for elem in vel_vec:
    df[elem+'_mean'] = df[elem].mean()
    df[[elem,elem+'_mean']].plot()
```
